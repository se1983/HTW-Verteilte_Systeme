```python

def generate_request_data(origin, destination, date):
    request_data = { 
        "request": {     
            "passengers": {
                "adultCount": 1},
            "slice": [{
                    "origin": origin,
                     "destination": destination,
                     "date": date}]}}
    return json.dumps(request_data)
```

```python
def do_request(origin, destination, date):

    get_response = requests.get(url=searchurl)
    post_data = generate_request_data(origin, destination, date)
    headers = {'Content-type': 'application/json'}
    response = requests.post(url=searchurl, data=post_data, headers=headers)
    
    return response


```

```python
from datetime import datetime

origin = 'MUC'
destinatioon = 'CDG'
date = datetime.today().strftime("%Y-%m-%d")

response = do_request(origin, destinatioon, date)
tripOptions = json.loads(response.content.decode('utf-8'))['trips']['tripOption']

def extract_tripdata(i, trip):
    return tripOptions[0]['slice'][0]['segment'][0]['leg'][0]
    

df = pd.DataFrame([extract_tripdata(i, trip) for i, trip in enumerate(tripOptions)])

df
```
